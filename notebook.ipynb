## Exercise 10: Convert column to datetime and extract month/year

**Purpose:**

Understand the distribution of property listings over time. Helps answer: When do most listings occur?

**Concepts Used:**

Pandas datetime conversion (pd.to_datetime)

Date attribute extraction (.dt.month, .dt.year)

```python
# Combine Yr Sold and Mo Sold into a datetime (first day of each month)
df["SaleDate"] = pd.to_datetime(
    df["Yr Sold"].astype(str) + "-" + df["Mo Sold"].astype(str) + "-01"
)

df["Sale_Year"] = df["SaleDate"].dt.year
df["Sale_Month"] = df["SaleDate"].dt.month

listings_over_time = df.groupby(["Sale_Year", "Sale_Month"]).size().reset_index(name="Count")
print(listings_over_time.head())
```

**Expected Output:**

A DataFrame snippet showing three columns:

Sale_Year

Sale_Month 

Count

**Business Insight:**

- Listings rise sharply from winter into spring/summer, with May showing peak activity (75 in 2006).

- This suggests that property sales are seasonal, with sellers preferring warmer months when buyer activity is higher.

- Real estate agents should focus marketing campaigns and open houses in spring and summer for maximum visibility.

- Developers and investors can time listings to align with these high-demand periods to secure better prices and faster sales.

## Exercise 15: Complex EDA â€“ Yearly House Price Trend (Line Chart)

**Purpose:**
 
To analyze how house prices have changed over the years. This reveals market trends (growth, decline, or stability) in real estate prices.

**Concepts Used:**

Pandas (groupby, mean)

Matplotlib (plot)

```python
# Calculate average price per year
yearly_trend = df.groupby("Yr Sold")["SalePrice"].mean()

# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(yearly_trend.index, yearly_trend.values, marker="o", linestyle="-", color="green")
plt.title("Yearly House Price Trend")
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

**Expected Output:**

A line chart showing average house price per year. Each point represents the mean property price in that year.

**Business Insight:**

- Peak average house prices occurred in 2007, followed by a sharp decline in 2008.

- A slight recovery happened in 2009, but prices dropped again to the lowest point in 2010.

- Overall, the trend from 2007 to 2010 shows a downward trajectory, likely reflecting the impact of the global financial crisis.

## Exercise 20 â€“ Heatmap of Average House Prices by Sale Year and Month

**Purpose:**

To analyze seasonal and yearly patterns in house prices, identifying which months and years tend to have higher or lower sale prices.

**Concepts Used:**

- Pandas groupby and unstack for pivoting data

- Matplotlib for visualization

- Seaborn heatmap for clear comparison 

```python
price_heatmap = df.groupby(['Yr Sold','Mo Sold'])['SalePrice'].mean().unstack()

plt.figure(figsize=(12,6))
sns.heatmap(price_heatmap, annot=True, fmt=".0f",cmap="YlGnBu")
plt.title("Average Price Heatmap (Year vs Month)")
plt.show()
```

**Expected Output:**

A heatmap where rows represent years sold and columns represent months sold. Each cell shows the average house price for that specific year-month combination. Colors indicate relative price levels (lighter = higher prices, darker = lower).

**Business Insights:**

- House prices show clear seasonal patterns, with *August and November* being the most profitable months for selling.

- Prices **peaked between 2006 and 2009**, then declined slightly in 2010.

- Buyers may find better deals in spring and early summer months like April, May, and July.

- Sellers should target *high-price months* to maximize returns, while buyers can focus on lower-price months for savings.
